```python
df["TypeOfSale"]=df.TypeOfSale.apply(lambda x : 1 if "residential_sale" in x else (2 if "monthly_rent" in x else None))



```

```python
df.dropna(subset=["TypeOfSale"],inplace=True)
df.shape
```

```python
print(df["StateOfBuilding"].unique())
print(df["PEB"].unique())
```

```python
x=df[df["TypeOfSale"]==1].groupby(by="PEB").Price.mean().index.values.tolist()
x.reverse()
```

```python
state_scaler= [None,"TO_BE_DONE_UP","TO_RESTORE","TO_RENOVATE","GOOD","JUST_RENOVATED","AS_NEW"]
peb_scaler=[None,'G_F',
 'G_C',
 'G',
 'F_E',
 'F_D',
 'F_C',
 'F',
 'E_D',
 'E_C',
 'E',
 'D',
 'C',
 'B_A',
 'B',
 'A'
 'A_A+',
 'A+'
 'A++',
 ]
```

```python
target=["Price"]
int_cols=["TypeOfProperty","TypeOfSale","BedroomCount","BathroomCount","ConstructionYear"]
bool_cols=["Garden","Terrace","SwimmingPool"]
float_cols=["LivingArea","GardenArea"]
ordinal=["StateOfBuilding","PEB"]
categorical=["SubtypeOfProperty","Kitchen","Province"]

columns_to_keep=target+int_cols+bool_cols+float_cols+ordinal+categorical
len(columns_to_keep)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

ct_int_cols=ColumnTransformer(("int_imput",SimpleImputer(strategy="most_frequent",copy=False)),remainder="passthrough")
ct_float_cols=ColumnTransformer(("imput_float",SimpleImputer(strategy="mean",copy=False)))
ct_categorical=ColumnTransformer(("cat_encoder",OneHotEncoder(handle_unknown="ignore")))

preprocessor=ColumnTransformer( transformers=
    [
        ("bool_col",SimpleImputer(strategy="constant",fill_value=0.0),bool_cols),
        ("int_col",SimpleImputer(strategy="most_frequent"),int_cols),
        ("float_cols",SimpleImputer(strategy="mean"),float_cols),
        ("ordinal_cols",OrdinalEncoder(categories=[state_scaler,peb_scaler],handle_unknown="use_encoded_value",unknown_value=-1),ordinal),
        ("categorical_col",OneHotEncoder(sparse_output=False),categorical)
    ],
    remainder="passthrough"
)
X=preprocessor.fit_transform(df[columns_to_keep])
df_preprocessed=pd.DataFrame(X,columns=preprocessor.get_feature_names_out())
df_preprocessed.head()
```

```python
preprocessor.feature_names_in_
```

```python
#Save Preprocessor
import pickle
pickle.dump(preprocessor,open("../data/preprocessor.pkl","wb"))
```

```python
y=df_preprocessed["remainder__Price"]
X=df_preprocessed.drop(["remainder__Price"],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=42)
print(X_train.shape,X_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error
score=mean_absolute_error(y_true=y_test,y_pred=model.predict(X_test))
print(score)

```
